```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
```

```python
from sklearn.datasets import make_blobs
```

```python
# Create Data
data = make_blobs(n_samples=400, n_features=5,centers=3,cluster_std=2,random_state=20)
```

```python
x=data[0][:,0]
y=data[0][:,1]
z=data[0][:,2]
```

```python
data[1]
```

```python
pd.DataFrame(list(data[0]))
```

```python
fig = plt.figure(figsize = (9, 9))
ax = plt.axes(projection ="3d")
ax.scatter(data[0][:,0],data[0][:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
centers=kmeans.cluster_centers_
labels=kmeans.labels_
```

```python
labels
```

```python
centers
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')#data after clustering it to 4 clusters
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')#dataset created for 3 clusters
```
