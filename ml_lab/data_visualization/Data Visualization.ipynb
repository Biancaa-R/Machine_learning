```python
titanic=pd.read_csv('titanic_train.csv')
ad=pd.read_csv('advertising.csv')
airports=pd.read_csv('airports.csv')
stud=pd.read_csv('StudentsPerformance.csv')
```

**Distribution plots**

Distribution of age of people on titanic

```python
plt.figure(figsize=(5,5))
sns.distplot(titanic['Age'],bins=40,kde=False)
```

Histogram of daily internet usage

```python
plt.figure(figsize=(5,5))
sns.histplot(ad['Daily Internet Usage'],bins=20,kde=True)
```

Jointplot of time spent on a website and the age of people

```python
plt.figure(figsize=(4,4))
sns.jointplot(x='Age',y='Daily Time Spent on Site',data=ad,kind='hex')
```

Pairplot for advertising data

```python
sns.pairplot(ad,hue='Clicked on Ad',palette='spring_r')
```

**Categorical plots**

```python
x=pd.get_dummies(stud['gender'])
```

```python
x.head()
```

```python
stud.drop('gender',axis=1,inplace=True)
stud['female']=x['female']
```

```python
stud.head()
```

Box and Whisker plot for students' math marks

```python
sns.boxplot(x="race/ethnicity", y="math score",data=stud ,palette='prism')
```

Violin plot for students' writing score

```python
sns.violinplot(x="race/ethnicity", y="writing score",data=stud, hue='female',split=True,palette='nipy_spectral_r')
```

Countplot for Embarked (from titanic dataset)

```python
sns.countplot(x='Pclass',data=titanic,hue='Survived',palette='Set1')
```

Barplot for dail internet usage of men

```python
sns.barplot(x='Male',y='Daily Internet Usage',data=ad,palette='plasma')
```

Pie chart to show no. of passengers and their location of mounting

```python
list(titanic['Embarked'].value_counts())
```

```python
loc=['Southampton','Cherbourg','Queenstown']
num=[644,168,77]
palette_color = sns.color_palette('rainbow')
plt.pie(num, labels=loc, colors=palette_color)
plt.show()
```

**Matrix plot**

Heatmap

```python
pv=flights.pivot_table(values='passengers',index='year',columns='month')
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(pv,cmap='magma',linecolor='white',linewidths=1)
```

**Other plots**

Line plot

```python
pv.head()
```

```python
plt.figure(figsize=(6,6))
sns.lineplot(data=pv,palette='hsv')
```

LM plot

```python
sns.lmplot(x='writing score',y='reading score',data=stud,hue='female',palette='Set1')
```
