```python
stud=pd.read_csv('StudentsPerformance.csv')
```

```python
stud.tail()
```

**Distribution plot**

```python
plt.figure(figsize=(5,5))
sns.distplot(stud['math score'],bins=40,kde=False)
```

**Histogram**

```python
plt.figure(figsize=(5,5))
sns.histplot(stud['math score'],bins=20,kde=True)
```

**Joint plot**

```python
plt.figure(figsize=(4,4))
sns.jointplot(x='writing score',y='reading score',data=stud,kind='hex')
```

**Pair plot**

```python
sns.pairplot(stud,hue='gender',palette='spring_r')
```

**Box Plot**

```python
plt.figure(figsize=(7,7))
sns.boxplot(x="race/ethnicity", y="math score",data=stud,palette='prism')
plt.show()
```

**Violin Plot**

```python
sns.violinplot(x="race/ethnicity", y="writing score",data=stud, hue='gender',split=True,palette='nipy_spectral_r')
```

**Count Plot**

```python
sns.countplot(x='race/ethnicity',data=stud,hue='gender',palette='Set1')
```

**Bar plot**

```python
sns.barplot(x='gender',y='math score',data=stud,palette='plasma')
```

**Pie chart**

```python
val=list(stud['race/ethnicity'].value_counts())
val
```

```python
lab=list(stud['race/ethnicity'].unique())
lab
```

```python
palette_color = sns.color_palette('rainbow')
plt.pie(val, labels=lab, colors=palette_color)
plt.show()
```

```python
from joypy import joyplot
```

```python
plt.figure()
joyplot(
    data=stud[['race/ethnicity','writing score']], 
    by='race/ethnicity',
    color='violet'
)
```

**LM plot**

```python
sns.lmplot(x='writing score',y='reading score',data=stud,hue='gender',palette='Set1')
```

**Line plot**

```python
plt.figure(figsize=(6,6))
sns.lineplot(data=stud.loc[0:10],palette='hsv')
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
sns.lineplot(data=stud['math score'].loc[0:10],palette='hsv')
plt.subplot(1,3,2)
sns.lineplot(data=stud['writing score'].loc[0:10],palette='hsv')
plt.subplot(1,3,3)
sns.lineplot(data=stud['reading score'].loc[0:10],palette='hsv')
```

**Scatter plot**

```python
sns.scatterplot(x='writing score',y='reading score',data=stud,hue='gender',palette='Set1')
```

**Heat map**

```python
sns.heatmap(stud[['math score','reading score','writing score']].loc[0:10])
```
